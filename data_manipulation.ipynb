# value_counts()
## Répondre à la question : combien de fois chaque valeur unique apparaît-elle dans une colonne ?

- e.g : quel est le nombre total d'élèves pour chaque session ?

```python
mention_counts = data['Session'].value_counts()
mention_counts
```

```python
mention_counts = data['Statut du candidat'].value_counts()
mention_counts
```

```python
mention_counts = data['Académie'].value_counts()
mention_counts
```

```python
mention_counts = data['Sexe'].value_counts()
mention_counts
```

## GroupBy

```python
mention_par_annee = data.groupby(['Session', 'Sexe', 'Voie']).size().unstack()
mention_par_annee
```

```python
mention_par_annee = data.groupby(['Session', 'Académie']).size().unstack()
mention_par_annee
```
